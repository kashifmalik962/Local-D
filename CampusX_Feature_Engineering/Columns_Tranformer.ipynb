```python
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer([
    ('trf1',SimpleImputer(),['engine_capacity']),
    ('trf2',OrdinalEncoder(categories=[['mechanical','automatic'],['gasoline','gas','diesel','hybrid-diesel','hybrid-petrol','electric'],
                                ['True','False'],['gasoline','diesel','electric'],['False','True'],['emergency','owned','new'],
                                ['True','False'],['False','True'],['False','True'],['False','True'],['False','True'],['False','True'],
                                ['False','True'],['False','True'],['False','True'],['False','True'],['False','True']],dtype=np.int32),
     ['transmission','engine_fuel','engine_has_gas','engine_type','has_warranty','state',
                          'is_exchangeable','feature_0','feature_1','feature_2','feature_3','feature_4',
                          'feature_5','feature_6','feature_7','feature_8','feature_9']),
    ('trf3',OneHotEncoder(drop='first',dtype=np.int32,sparse=False),['color','body_type','drivetrain','location_region'])
],remainder='passthrough')
```

```python
transform_data = transformer.fit_transform(data)
```

```python
manu_counts = data['manufacturer_name'].value_counts()
```

```python
model_counts = data['model_name'].value_counts()
```

```python
repl_manu = manu_counts[manu_counts <= 200].keys()
```

```python
repl_model = model_counts[model_counts <= 150].keys()
```

```python
repl_model
```

```python
manu_dummy = pd.get_dummies(data['manufacturer_name'].replace(repl_manu, 'uncommon_manufacture'), drop_first=True)
```

```python
model_dummy = pd.get_dummies(data['model_name'].replace(repl_model, 'uncommon_model'),drop_first=True)
```

```python
dummy_df = pd.concat([manu_dummy,model_dummy], axis=1)
```

```python
new_data = np.concatenate([dummy_df.values,transform_data], axis=1)
```

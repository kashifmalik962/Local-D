```python
import numpy as np
import pandas as pd

import scipy.stats as stats
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean().round(),inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(121)
sns.distplot(X_train['Fare'])
plt.title('Fare PDF')


plt.subplot(122)
stats.probplot(X_train['Fare'], dist='norm', plot=plt)
plt.title('Fare QQ Plot')

plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(121)
sns.distplot(X_train['Age'])
plt.title('Age PDF')

plt.subplot(122)
stats.probplot(X_train['Age'], dist='norm', plot=plt)
plt.title('Fare QQ Plot')
plt.show()
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
```

```python
cross_val_score(clf1,X_test,y_test).mean()
```

```python
cross_val_score(clf2,X_test,y_test).mean()
```

```python
trf = FunctionTransformer(func=np.log1p)
```

```python
X_train_transformed = trf.fit_transform(X_train)
X_test_transformed = trf.transform(X_test)
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf1.fit(X_train_transformed, y_train)
clf2.fit(X_train_transformed, y_train)
```

```python
cross_val_score(clf1, X_test_transformed,y_test).mean()
```

```python
cross_val_score(clf2, X_test_transformed,y_test).mean()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
stats.probplot(X_train['Fare'], dist='norm',plot=plt)
plt.title('Fare_Dist Before Log')

plt.subplot(122)
stats.probplot(X_train_transformed['Fare'], dist='norm', plot=plt)
plt.title('Fare_Dist After Log')
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
stats.probplot(X_train['Age'], dist='norm',plot=plt)
plt.title('Age_Dist Before Log')

plt.subplot(122)
stats.probplot(X_train_transformed['Age'], dist='norm', plot=plt)
plt.title('Age_Dist After Log')
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X_train['Age'])
plt.title('Before log')

plt.subplot(122)
sns.distplot(X_train_transformed['Age'])
plt.title('after log')
plt.show()
```

```python
trf2 = ColumnTransformer([
    ('log',FunctionTransformer(np.log1p),['Fare'])
],remainder='passthrough')
```

```python
X_train_transformed2 = trf2.fit_transform(X_train)
X_test_transformed2 = trf2.transform(X_test)
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf1.fit(X_train_tranformed2,y_train)
clf2.fit(X_train_tranformed2,y_train)
```

```python
cross_val_score(clf1,X_test_tranformed2,y_test).mean()
```

```python
cross_val_score(clf2,X_test_tranformed2,y_test).mean()
```

```python
def apply_transform(transform):
    X = df.iloc[:,0:2]
    y = df.iloc[:,-1]
    
    trf = ColumnTransformer([('log',FunctionTransformer(transform),['Fare'])],remainder='passthrough')
    
    X_trains = trf.fit_transform(X)   
    
    clff = LogisticRegression()
    
    print("Accuracy Score :", cross_val_score(clff, X_trains,y).mean())
    
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    stats.probplot(X['Fare'], dist='norm',plot=plt)
    plt.title('Before transform')
    
    
    plt.subplot(122)
    stats.probplot(X_trains[:,0], dist='norm',plot=plt)
    plt.title('After transform')
    plt.show()
```

```python
apply_transform(lambda x: x**3)
```

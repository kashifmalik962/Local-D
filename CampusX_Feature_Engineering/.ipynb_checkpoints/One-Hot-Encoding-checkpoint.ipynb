```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df.head(6)
```

```python
ohe_data_cols = df['fuel'].unique()
```

```python
ohe_data_cols
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False,dtype=np.int32)
```

```python
ohe_data = ohe.fit_transform(df[['fuel']])
```

```python
ohe_data
```

```python
ohe_data_new = pd.DataFrame(ohe_data, columns=ohe_data_cols)
```

```python
df1 = pd.concat([df,ohe_data_new], axis=1)
```

```python
df1.drop(['CNG','fuel'], axis=1,inplace=True)
```

```python
df1.head(6)
```

```python
keys_owner = df1['owner'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Test Drive Car','Fourth & Above Owner','Third Owner','Second Owner','First Owner']],dtype=np.int32)
```

```python
oe_data = oe.fit_transform(df1[['owner']])
```

```python
oe_data_repl = pd.DataFrame(oe_data,columns=['owner_cat'])
```

```python
df2 = pd.concat([df1,oe_data_repl],axis=1)
```

```python
df2.drop('owner',axis=1,inplace=True)
```

```python
df2.head(6)
```

```python
counts = df2['brand'].value_counts()
```

```python
number = 100
```

```python
under_100 = counts[counts <=100].keys()
under_100
```

```python
brand_dummy = pd.get_dummies(df2['brand'].replace(under_100, 'uncommon'))
```

```python
df3 = pd.concat([brand_dummy,df2], axis=1)
```

```python
df3.drop('brand',axis=1, inplace=True)
```

```python
df3
```

```python
sel_price = df[['selling_price']]
```

```python
df = pd.concat([df3,sel_price], axis=1)
```

## Feature Engineering

### Data Cleaning

```python
# Divide Data into Groups
df1.groupby('area_type')['area_type'].agg('count')
```

```python
# Drop unnecessary columns
df2 = df1.drop(['area_type','society','balcony','availability'],axis=1)
df2.head()
```

```python
# Remove null values
df3 = df2.dropna()
df3.isnull().sum()
```

```python
# Analyse how many unique values have a particular column
df3['size'].unique()
```

```python
# Create a New columns from another columns from transformated data
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))
df3.head()
```

```python
# Check data is greater then given value
df3[df3['bhk'] > 20] 
```

```python
# Check how many unique values in a column
df3.total_sqft.unique()
```

```python
df3.loc[648]
```

```python
# Create a function
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True 
```

```python
# The ~ operator is then used to negate the boolean mask
df3[~df3['total_sqft'].apply(is_float)].head(10)
```

```python
# Native Function
def convert_into_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        None
```

```python
convert_into_num('34.46Sq. Meter')
```

```python
# Create a copy of DataFrame
df4 = df3.copy()
```

```python
# Apply a Native Function
df4['total_sqft'] = df4['total_sqft'].apply(convert_into_num)
df4.head()
```

```python
# Create a deep copy of DataFrame
df5 = df4.copy()
df5.head()
```

```python
# Create a column per square foot Area
df5['per_sqft_price'] = (df5['price']*100000/df5['total_sqft']).round(4)
df5.head()
```

```python
# How many unique values in a column
df5['location'].unique()
```

```python
# remove extra space all values in a columns that is a good practice before apply group-by Function
df5['location'] = df5['location'].apply(lambda x: x.strip())
```

```python
# Groups the location column and count how many values concerned with their group
location_group = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_group
```

```python
# Find the group that have less then 10 values
location_less_then_10 = location_group[location_group<=10]
```

```python
# diffrentiate the values there are 241 values that have values greater than 10
len(location_group)-len(location_group[location_group<=10])
```

```python
# we are set the thresold by a name 'other' that have record less than 10y the lambda function 
df5['location'] = df5['location'].apply(lambda x: 'other' if x in location_less_then_10 else x)
df5['location'].nunique()
```

### Outlier Detect and Removal

```python
# 1 BHK require almost 300 total_sqft area we are filtering 1 BHK less than 300 total_sqft and remove
df5[df5['total_sqft']/df5['bhk'] < 300]
```

```python
# Using Negate operation
df6 = df5[~(df5['total_sqft']/df5['bhk'] < 300)]
df6.head()
```

```python
# Remove Outlier for far away 1 Beyond std
def remove_outlier(df,column):
    mean_df = df[column].mean()
    std_df = df[column].std()
    reduse_df = df[(df[column] <= (mean_df+std_df)) & (df[column] > (mean_df-std_df))]
    return reduse_df

df7 = remove_outlier(df6,'per_sqft_price')
df7.shape
```

```python
# diffrentiate 2 & 3 BHK and Plot a particular Location
def scatter_plot(df,location):
    plt.figure(figsize=(7, 5))
    bhk2 = df[(df['location']==location)  & (df['bhk']==2)]
    bhk3 = df[(df['location']==location)  & (df['bhk']==3)]
    plt.scatter(bhk2['total_sqft'],bhk2['price'], marker='.',color='blue',label='bhk2',s=50)
    plt.scatter(bhk3['total_sqft'],bhk3['price'], marker='+',color='green',label='bhk3',s=50)
    plt.title(location)
    plt.legend(loc='upper left')

scatter_plot(df7,'Hebbal')
```

```python
# We should also remove properties where for same location, the price of (for example) 3 bedroom apartment is less than 2 bedroom apartment 
# (with same square ft area). What we will do is for a given location, we will build a dictionary of stats per bhk
def remove_outlier_bhk(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df['per_sqft_price']),
                'std' : np.std(bhk_df['per_sqft_price']),
                'count' : bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df['per_sqft_price']<(stats['mean'])].index.values)
                
    return df.drop(exclude_indices,axis='index')


df8 = remove_outlier_bhk(df7)
df8.shape
```

```python
# Plot a histogram
plt.hist(df8['per_sqft_price'],rwidth=0.8)
```

```python
# Plot a histogram
plt.hist(df8['bath'],rwidth=0.8)
```

```python
# Remove the record which have bathroom is more than 2 bhk
df9 = df8[~(df8['bath'] > df8['bhk']+2)]
df9.head()
```

```python
# We are Using OneHotEncoding on Location Column/variable
ohe = OneHotEncoder(sparse=False,handle_unknown='ignore',dtype='int32')

# Fit and transform the data
ohe_encoded = ohe.fit_transform(df9[['location']])
ohe_data = pd.DataFrame(ohe_encoded, columns=ohe.get_feature_names_out(['location'])) 
ohe_data.head(4)
```

```python
# Reset index is a good practice before concat diffrent DataFrames
df9 = df9.reset_index()
ohe_data = ohe_data.reset_index()
```

```python
# Concat two diffrent Dataframe
df10 = pd.concat([df9,ohe_data],axis='columns')
df10.head(4)
```

```python
# Drop unnessasary columns
df11 = df10.drop(['index','location','size','per_sqft_price'],axis=1)
df11.head(4)
```

```python
# Break the Data into Input Data and Target Data
X = df11.drop('price',axis=1)
y = df11.price
```

```python
# Split Data into Training and Testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=10)
```

```python
# Import LinearRegression Class
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
# Cross-Validation Score 
from sklearn.model_selection import cross_val_score
cv = cross_val_score(LinearRegression(),X,y,cv=5)
np.mean(cv)
```

```python
# Select the best Model/parameter to fit our Data
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(x,y):
    algos = {
        'linear regression': {
            'model' : LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=5, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score' : gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame (scores,columns=['model', 'best_score', 'best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
# Pridict Output
def predict_price(total_sqft, bath, bhk, location):
    
    if location in X.columns:
        # print(123)
        loc_index = np.where(X.columns==location)[0][0]
        x = np. zeros(len(X.columns))
        x[0] = total_sqft
        x[1] = bath
        x[2] = bhk
        x[loc_index] = 1
        return lr_clf.predict([x])[0]
    else:
        print('Invalid Location')

predict_price(1056,2,2,'location_1st Block Jayanagar')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('IRIS.csv')
iris.head()
```

```python
print(iris.shape)
```

```python
print(iris.columns)
```

```python
# how many flowers for each species are present
iris['species'].value_counts()
```

## 2-D scatter plot

```python
iris.plot(kind='scatter',x='sepal_length',y='sepal_width')
```

```python
plt.title('Scatter Plot with Legends')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid()

plt.scatter(x=iris.sepal_length,y=iris.sepal_width,color='blue')
plt.show()
```

```python

plt.title('Scatter Plot with Legends')
plt.grid()
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width,hue=iris.species)
plt.show()
```

## 3D Scatter plot

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(iris.sepal_length,iris.sepal_width,iris.sepal_length, c='b', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_length', color='species')
fig.update_traces(marker=dict(size=4, opacity=0.8))
fig.show()
```

## (3) Pair Plot

```python
plt.close();
sns.set_style('whitegrid')
sns.pairplot(iris,hue='species',size=3)
plt.show()
```

## (4) Histogram

```python
plt.hist(iris.petal_length)
```

```python
sns.FacetGrid(iris, hue='species',height=6).map(sns.distplot,'petal_length').add_legend()
```

## boxplot

```python
sns.boxplot(x = iris.species ,y = iris.petal_length)
```

## Violin plot

### A Violin plot combines the benefits of the previous plot 

```python
sns.violinplot(x=iris.species, y=iris.petal_length,data=iris)
plt.show()
```

```python
sns.violinplot(iris.petal_length)
```

```python
sns.displot(iris.petal_length,kde=True)
```

```python
sns.distplot(iris.petal_length)
```

```python
setosa = iris[iris['species']=='Iris-setosa']
setosa
```

```python
sns.jointplot(x = setosa.petal_length, y = setosa.petal_width,data = setosa, kind='kde',color='blue')
plt.show()
```
